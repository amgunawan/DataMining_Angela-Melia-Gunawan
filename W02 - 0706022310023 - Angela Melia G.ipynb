### **Handle Missing Values**

```python
# Identify missing values
# Calculate null values and percentage
null_counts = df.isnull().sum()
null_percentages = (null_counts / len(df)) * 100

# Create a DataFrame for columns with null values
null_df = pd.DataFrame({'Null Count': null_counts, 'Null Percentage': null_percentages})

# Filter for columns with null values
null_df = null_df[null_df['Null Count'] > 0]

# Format the percentage with two decimal places and % symbol
null_df['Null Percentage'] = null_df['Null Percentage'].map('{:.2f}%'.format)


null_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the horizontal bar plot
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
sns.barplot(x='Null Count', y=null_df.index, data=null_df, orient='h', palette='rainbow')
plt.xlabel('Number of Null Values', fontsize=12)
plt.ylabel('Columns', fontsize=12)
plt.title('Null Values per Column', fontsize=14)

# Add labels to the bars
for i, v in enumerate(null_df['Null Count']):
    plt.text(v + 1, i, str(v), color='black', va='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Display unique values for each column
for col in df.columns:
  print(f"Unique values for column '{col}':")
  print(df[col].unique())
  print("-" * 20)
```

```python
# Drop 'url' and 'title' column
# Dilakukan karena data kurang berguna untuk keperluan analisis
df = df.drop('url', axis=1, errors='ignore')
df = df.drop('title', axis=1, errors='ignore')

# Drop 'address' column
# Dilakukan karena merupakan kombinasi dari kolom 'district' dan 'city'
df = df.drop('address', axis=1, errors='ignore')

# Drop 'property_type' column
# Dilakukan karena seluruh data merupakan data properti rumah
df = df.drop('property_type', axis=1, errors='ignore')

# Drop 'building_orientation' column
# Dilakukan karena terlalu banyak missing values hingga 46%
df = df.drop('building_orientation', axis=1, errors='ignore')

# Drop 'year_built' column
# Dilakukan karena dari kolom 'building_age' bisa tahu kapan year built, serta value dari 'year_built' ada yang
df = df.drop('year_built', axis=1, errors='ignore')

df.head()
```

```python
# Handle cases where the value of 'ads_id' and 'certificate' columns are null
df['ads_id'] = df['ads_id'].fillna('unknown')
df['certificate'] = df['certificate'].fillna('unknown')
```

```python
# Imputation data for 'building_size_m2', 'land_size_m2', and 'floors' column
# Karena terdapat outlier pada kolom 'building_size_m2' dan 'land_size_m2', sehingga jika menggunakan mean, maka nilai hasil imputasi bisa jadi bias.
# Dengan median, nilainya tetap representatif untuk mayoritas data.
df.fillna({
    'building_size_m2': df['building_size_m2'].median(),
    'land_size_m2': df['land_size_m2'].median(),
    'floors': df['floors'].median()
}, inplace=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
import numpy as np

# Bedrooms memiliki hubungan kuat dengan kolom berikut, sehingga kolom berikut digunakan
features_bed = ['building_size_m2', 'land_size_m2', 'floors']

mask_bed_missing = df['bedrooms'].isna()

# Train & Test
train_bed = df[df['bedrooms'].notna()]
test_bed = df[df['bedrooms'].isna()]

# Train Model & Predict
if not test_bed.empty:
  model_bed = DecisionTreeRegressor(random_state=42)
  model_bed.fit(train_bed[features_bed], train_bed['bedrooms'])
  pred = model_bed.predict(test_bed[features_bed])
  df.loc[df['bedrooms'].isna(), 'bedrooms'] = np.round(pred).astype(int)

  print(df.loc[mask_bed_missing, ['bedrooms', 'building_size_m2', 'land_size_m2', 'floors']])
```

Decision Tree dipilih karena fleksibel untuk mempelajari pola non-linear, tidak butuh pre-processing rumit, mudah dijelaskan, dan cocok untuk memprediksi jumlah kamar tidur berdasarkan ukuran bangunan, tanah, dan lantai.

```python
# Bathrooms memiliki hubungan kuat dengan kolom berikut, sehingga kolom berikut digunakan
features_bed = ['building_size_m2', 'land_size_m2', 'floors', 'bedrooms']

mask_bath_missing = df['bathrooms'].isna()

# Train & Test
train_bed = df[df['bathrooms'].notna()]
test_bed = df[df['bathrooms'].isna()]

# Train Model & Predict
if not test_bed.empty:
  model_bed = DecisionTreeRegressor(random_state=42)
  model_bed.fit(train_bed[features_bed], train_bed['bathrooms'])
  pred = model_bed.predict(test_bed[features_bed])
  df.loc[df['bathrooms'].isna(), 'bathrooms'] = np.round(pred).astype(int)

  print(df.loc[mask_bed_missing, ['bathrooms', 'building_size_m2', 'land_size_m2', 'floors', 'bedrooms']])
```

Decision Tree digunakan karena mampu memanfaatkan hubungan logis antara ukuran bangunan, luas tanah, jumlah lantai, dan jumlah kamar tidur untuk memperkirakan jumlah kamar mandi, tanpa perlu pre-processing rumit dan tetap memberikan hasil yang mudah dijelaskan.

```python
# Identifikasi baris yang salah
mask = df["property_condition"].isin(["semi furnished", "unfurnished"])

# Simpan sementara nilai yang salah
temp = df.loc[mask, "property_condition"]

# Tukar isi kolom
# Isi kolom 'property_condition' dengan kolom 'furnishing'
df.loc[mask, "property_condition"] = df.loc[mask, "furnishing"]

# Isi kolom 'furnishing' dengan nilai yang tadinya salah
df.loc[mask, "furnishing"] = temp
```

Beberapa nilai unik pada kolom property_condition dan furnishing ternyata ada yang tertukar. Misalnya:
*   Nilai seperti "semi furnished" atau "unfurnished" seharusnya masuk ke kolom furnishing, tapi malah tercatat di kolom property_condition.
*   Sebaliknya, kolom furnishing jadi kehilangan nilai yang benar karena posisi data tertukar.

Sehingga, dilakukan langkah seperti di atas.

```python
# Fill the missing values of 'property_condition' and 'furnishing' column
df['property_condition'] = df['property_condition'].fillna(df['property_condition'].mode()[0])
df['furnishing'] = df['furnishing'].fillna(df['furnishing'].mode()[0])
```

### **Feature Engineering**

```python
# Change data types from float to integer
df['bedrooms'] = df['bedrooms'].astype(int)
df['bathrooms'] = df['bathrooms'].astype(int)
df['land_size_m2'] = df['land_size_m2'].astype(int)
df['building_size_m2'] = df['building_size_m2'].astype(int)
df['carports'] = df['carports'].astype(int)
df['maid_bedrooms'] = df['bedrooms'].astype(int)
df['maid_bathrooms'] = df['bathrooms'].astype(int)
df['floors'] = df['floors'].astype(int)
df['garages'] = df['garages'].astype(int)
```

```python
# 1. Normalize the format
# Dilakukan karena terdapat beberapa data yang spasinya lebih dari 1 setelah koma
df['facilities'] = df['facilities'].str.replace(r"\s*,\s*", ", ", regex=True) # trim commas
df['facilities'] = df['facilities'].str.replace(r"\s{2,}", " ", regex=True) # remove excess spaces

# 2. Change into a list of facilities
df['facilities_list'] = df['facilities'].str.split(", ")

# 3. Count the number of facilities
# Isi kolom 'facilities' diubah untuk mempermudah pengolahan data karena berupa data numerik
df['facilities'] = df['facilities_list'].apply(len)

# 4. Drop 'facilities' and 'facilities_list' column
df = df.drop('facilities_list', axis=1, errors='ignore')
```

```python
# Define bins dan labels
bins = [0, 5, 10, 15, 20, df['facilities'].max()]
labels = ['1-5', '6-10', '11-15', '16-20', '>20']

# Binning
df['facilities'] = pd.cut(
    df['facilities'],
    bins=bins,
    labels=labels,
    include_lowest=True
)

# Preview
print(df['facilities'].value_counts())
```

```python
# Define bins and labels
bins = [0, 5, 10, 15, df['building_age'].max()]
labels = ['0-5 years', '6-10 years', '11-15 years', '16-20 years']

# First, bin only numeric values
df['building_age'] = pd.cut(
    df['building_age'],
    bins=bins,
    labels=labels,
    include_lowest=True
)

# Add "unknown" category and fill NaN
df['building_age'] = df['building_age'].cat.add_categories('unknown')
df['building_age'] = df['building_age'].fillna('unknown')

# Preview
print(df['building_age'].value_counts())
```
